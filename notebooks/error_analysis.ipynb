```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cm = np.array([
[976,1,1,2,0,2,1,0,2,1],
[0,1116,1,2,1,1,0,2,1,1],
[1,5,968,10,3,0,0,5,6,1],
[1,1,1,1001,0,8,0,2,6,0],
[0,0,0,0,949,0,2,0,1,23],
[3,1,0,5,1,881,5,0,1,5],
[4,0,0,0,3,4,970,0,1,0],
[2,3,2,2,5,0,0,1014,1,13],
[1,4,1,7,0,4,3,0,948,7],
[0,3,0,3,9,1,0,4,3,971]
])

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for Best Model")

plt.show()
plt.savefig("confusion_matrix.png")
```

```python
misclassified = cm.copy()
np.fill_diagonal(misclassified, 0)

plt.figure(figsize=(8,6))
sns.heatmap(misclassified, annot=True, fmt="d", cmap="Reds")

plt.title("Misclassification Patterns")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()
plt.savefig("misclassification_matrix.png")
```
